```python
import os
def find_file(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)
    return None

# 自动搜索底模路径
g_path = find_file("s2Gv2Pro.pth", "/content/drive/MyDrive")
d_path = find_file("s2Dv2Pro.pth", "/content/drive/MyDrive")

if g_path and d_path:
    print(f"✅ 找到底模了！\nGPT路径: {g_path}\nSoVITS路径: {d_path}")
    # 存入变量供下一步使用
    %store g_path
    %store d_path
else:
    print("❌ 在 Drive 里没找到底模，请确认你是否已经把 s2Gv2Pro.pth 上传到了 Google Drive。")
```
